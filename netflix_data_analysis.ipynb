# Netflix Data Analysis with Python

This project analyzes a public Netflix dataset using Python, pandas, and matplotlib.  
The goal is to extract useful insights about content types, countries of origin, release years, and the most common genres.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df["type"].value_counts().plot(kind="bar", color=["red", "blue"])
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df["country"].value_counts().head(10).plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Countries with Most Content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
df["release_year"].value_counts().sort_index().plot(kind="bar", figsize=(15,5))
plt.title("Number of Releases by Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
df["listed_in"].value_counts().head(10).plot(kind="bar", figsize=(10,5))
plt.title("Most Common Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

## Conclusions

- There are more movies than TV shows in the Netflix catalog.
- The United States and India produce the most content.
- The number of new releases increased significantly after 2017, with a peak in 2019.
- The most common genres are documentaries, dramas, and comedies.
